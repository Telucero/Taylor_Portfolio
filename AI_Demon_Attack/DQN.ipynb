```python
## This needs to be fixed and updated
# module or custom set formatting

class DQN(torch.nn.Module):
    def __init__(self,observation):
        super(DQN,self).__init__()
        self.net= torch.nn.Sequential(
            torch.nn.Linear(128,768),
            torch.nn.ReLU(),
            torch.nn.Linear(768, 1536),
            torch.nn.ReLU(),
            torch.nn.Linear(1536, 6),
            #torch.nn.Softmax()
            #torch.nn.Flatten(256, 1)

        ).to(device)
        
    def forward(self, inputs):
        #print(inputs)
        inputs = self.net(inputs)
        new_shape = (len(inputs))
        inputs = inputs.view(new_shape)
        return inputs
        
        
```

```python
Experience = namedtuple(
    'Experience',
    ('observation', 'action', 'reward', 'pre_observation', 'diff_observation')
)
```

```python
def pull_tensors(exp):
    memory = []
    experience_tracked = 0 
    storage = 100000
    
    if len(memory) < storage:
            memory.append(exp)
    else:
        memory[experience_tracked % storage] = exp
        experience_tracked += 1
        
    for x in exp:      
        batch = Experience(*zip(x))
        #print(len(batch))
        #print("this",batch) 

        tensor1= torch.cat(batch.observation)
        #print(batch.action[0].view(-1))
        tensor2= torch.cat((batch.action[0].view(-1), torch.zeros(5)))
        tensor3= torch.cat((batch.reward[0].view(-1), torch.zeros(5)))
        tensor4= torch.cat(batch.pre_observation)
        tensor5= torch.cat(batch.diff_observation)
    
    return (tensor1, tensor2, tensor3, tensor4, tensor5)
```


The learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.
For reinforcement learning it refers to how much the model learns. If for example the model has an alpha of 0 then the Q-values wonât update and nothing is learned, however, on the other hand if alpha is set to 0.9 then it will learn rapidly.  It terms of this program of Demon attack, changing the learning rate improved how fast it learned a specific move. On one when firing it fired very well,  going up to scores of 940 to 1635, although the agent itself doesnât move. If initial, movements are a change in lateral position then the agent has tendency to move right or left in attempts to avoid shooting. With a lower rate at about .01  it takes more episodes for the agent to improve and maximize rewards. I tried implementing a punishment whenever the agent dies providing a negative reward in an attempt to help the agent learn to avoid the lasers. With the alpha trying to minimize the loss function of the smooth l1 loss certain episodes will drastically especially when the learning is high, this because it takes a larger step downward on the curve to minimize the gradient. A larger learning however has the downside of causing divergence faster, as the step could over shoot the minimum. With DQN a single network is not as stable, hence the division of the policy network and the target network. The target network focuses Q(sâ, aâ) in the bellman equation. The second Q-values provided by this network are used to back propagate through and train the main network. This is to improve stability of the network.
